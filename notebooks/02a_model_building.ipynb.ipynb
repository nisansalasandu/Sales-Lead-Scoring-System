```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('../data/cleaned_sales_leads_dataset.csv')
```

```python
# One-hot encode selected columns
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
from imblearn.over_sampling import SMOTE

# Use the encoded dataframe for SMOTE
X_encoded = df_encoded.drop('converted', axis=1)
y_encoded = df_encoded['converted']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_encoded, y_encoded)

print(f"Original dataset: {len(y_encoded)} samples")
print(f"Resampled dataset: {len(y_resampled)} samples")
print(f"\nClass distribution after SMOTE:")
print(y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled,
    y_resampled,
    test_size=0.2,
    random_state=42,
    stratify=y_resampled
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# Build and train both models (data is already encoded, so no preprocessing needed)

# 1. Logistic Regression Model
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)

# 2. Decision Tree Model
dec_tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)

# Train both models
log_reg_model.fit(X_train, y_train)
dec_tree_model.fit(X_train, y_train)

print("Both models have been trained successfully.")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Comparison and Evaluation

# Get predictions
log_preds = log_reg_model.predict(X_test)
tree_preds = dec_tree_model.predict(X_test)

# Calculate Accuracy
log_acc = accuracy_score(y_test, log_preds)
tree_acc = accuracy_score(y_test, tree_preds)

print(f"Logistic Regression Accuracy: {log_acc:.2%}")
print(f"Decision Tree Accuracy: {tree_acc:.2%}")

# Show detailed reports for both models
print("\n" + "="*50)
print("Logistic Regression Report:")
print("="*50)
print(classification_report(y_test, log_preds))

print("\n" + "="*50)
print("Decision Tree Report:")
print("="*50)
print(classification_report(y_test, tree_preds))
```

```python
import joblib

# Selection and Saving the Best Model
# Selection logic
if log_acc > tree_acc:
    best_model = log_reg_model
    model_name = "Logistic Regression"
else:
    best_model = dec_tree_model
    model_name = "Decision Tree"

print(f"Winning Model: {model_name}")

# Save the best model for the scoring system
joblib.dump(best_model, '../models/best_lead_scoring_model 1.pkl')
print(f"The {model_name} model has been saved as 'best_lead_scoring_model.pkl'")
```
