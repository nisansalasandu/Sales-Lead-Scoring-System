```python
# Import Libraries
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 10)

print("Libraries imported successfully!")
```

```python
# Load Scored data

print("\n Loading Scored Data")
print("-"*70)

# Load scored data from notebook 03
df = pd.read_csv('../data/scored_leads.csv')

print(f" Loaded {len(df)} scored leads")
print(f" Columns: {df.shape[1]}")
print(f" Score range: {df['lead_score'].min():.2f} - {df['lead_score'].max():.2f}")
```

```python
# Sales Funnel Analysis

print("\n" + "="*70)
print("Sales Funnel Analysis")
print("-"*70)

# Define funnel stages in order
stage_order = ['New', 'Engaged', 'Qualified', 'Demo/Trial', 'Converted']

# Calculate metrics for each stage
funnel_metrics = []

for i, stage in enumerate(stage_order):
    stage_leads = df[df['pipeline_stage'] == stage]
    
    metrics = {
        'stage': stage,
        'stage_number': i + 1,
        'total_leads': len(stage_leads),
        'converted_leads': stage_leads['converted'].sum(),
        'conversion_rate': (stage_leads['converted'].mean() * 100) if len(stage_leads) > 0 else 0,
        'avg_lead_score': stage_leads['lead_score'].mean() if len(stage_leads) > 0 else 0,
        'high_priority': len(stage_leads[stage_leads['priority'] == 'HIGH']),
        'medium_priority': len(stage_leads[stage_leads['priority'] == 'MEDIUM']),
        'low_priority': len(stage_leads[stage_leads['priority'] == 'LOW']),
        'avg_engagement': stage_leads['engagement_score'].mean() if len(stage_leads) > 0 else 0
    }
    funnel_metrics.append(metrics)

funnel_df = pd.DataFrame(funnel_metrics)

# Calculate drop-off rates
funnel_df['drop_off_rate'] = 0.0
for i in range(1, len(funnel_df)):
    prev_leads = funnel_df.loc[i-1, 'total_leads']
    curr_leads = funnel_df.loc[i, 'total_leads']
    if prev_leads > 0:
        funnel_df.loc[i, 'drop_off_rate'] = ((prev_leads - curr_leads) / prev_leads) * 100

print("Funnel metrics calculated")
print("\n Funnel Summary:")
print(funnel_df.to_string(index=False))

# Save funnel metrics
funnel_df.to_csv('../reports/funnel_metrics.csv', index=False)
print("\n Saved: reports/funnel_metrics.csv")
```

```python
# Advanced Visualizations

print("\n" + "="*70)
print("Creating Advanced Visualizations")
print("-"*70)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Sales Funnel
ax1 = axes[0, 0]
funnel_colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A']
bars = ax1.barh(funnel_df['stage'], funnel_df['total_leads'], color=funnel_colors)
ax1.set_title('Sales Pipeline Funnel - Lead Progression', fontsize=14, fontweight='bold')
ax1.set_xlabel('Number of Leads')
ax1.set_ylabel('Funnel Stage')
for i, (bar, value) in enumerate(zip(bars, funnel_df['total_leads'])):
    ax1.text(value + 10, i, str(value), va='center', fontweight='bold')

# 2. Conversion Rate by Stage
ax2 = axes[0, 1]
bars = ax2.bar(funnel_df['stage'], funnel_df['conversion_rate'], color=funnel_colors)
ax2.set_title('Conversion Rate by Pipeline Stage', fontsize=14, fontweight='bold')
ax2.set_xlabel('Pipeline Stage')
ax2.set_ylabel('Conversion Rate (%)')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
for bar, value in zip(bars, funnel_df['conversion_rate']):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{value:.1f}%', ha='center', va='bottom', fontweight='bold')

# 3. Priority Distribution by Stage
ax3 = axes[1, 0]
stages = funnel_df['stage']
x = np.arange(len(stages))
width = 0.25

ax3.bar(x - width, funnel_df['high_priority'], width, label='HIGH', color='#00CC96')
ax3.bar(x, funnel_df['medium_priority'], width, label='MEDIUM', color='#FFA15A')
ax3.bar(x + width, funnel_df['low_priority'], width, label='LOW', color='#EF553B')

ax3.set_title('Priority Distribution Across Pipeline Stages', fontsize=14, fontweight='bold')
ax3.set_xlabel('Pipeline Stage')
ax3.set_ylabel('Number of Leads')
ax3.set_xticks(x)
ax3.set_xticklabels(stages, rotation=45, ha='right')
ax3.legend()

# 4. Drop-off Analysis
ax4 = axes[1, 1]
stages_with_dropoff = funnel_df[funnel_df['drop_off_rate'] > 0]
ax4.plot(stages_with_dropoff['stage'], stages_with_dropoff['drop_off_rate'], 
         marker='o', markersize=10, linewidth=3, color='red')
ax4.set_title('Lead Drop-off Rate Between Stages', fontsize=14, fontweight='bold')
ax4.set_xlabel('Pipeline Stage')
ax4.set_ylabel('Drop-off Rate (%)')
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=45, ha='right')
ax4.grid(True, alpha=0.3)
for x, y in zip(stages_with_dropoff['stage'], stages_with_dropoff['drop_off_rate']):
    ax4.text(x, y + 2, f'{y:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../reports/sales_funnel_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved: reports/sales_funnel_visualization.png")
```

```python
# Industry and Location Analysis

print("\n" + "="*70)
print("Industry & Location Analysis")
print("-"*70)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Conversion rate by industry
ax1 = axes[0]
industry_conv = df.groupby('industry').agg({
    'converted': ['sum', 'count', 'mean'],
    'lead_score': 'mean'
}).round(2)
industry_conv.columns = ['converted_count', 'total_leads', 'conversion_rate', 'avg_score']
industry_conv['conversion_rate'] = industry_conv['conversion_rate'] * 100
industry_conv = industry_conv.sort_values('conversion_rate', ascending=False)

bars = ax1.barh(industry_conv.index, industry_conv['conversion_rate'], color='teal')
ax1.set_title('Conversion Rate by Industry', fontsize=14, fontweight='bold')
ax1.set_xlabel('Conversion Rate (%)')
for i, (bar, value) in enumerate(zip(bars, industry_conv['conversion_rate'])):
    ax1.text(value + 1, i, f'{value:.1f}%', va='center')

# Conversion rate by location (top 10)
ax2 = axes[1]
location_conv = df.groupby('location').agg({
    'converted': ['sum', 'count', 'mean'],
    'lead_score': 'mean'
}).round(2)
location_conv.columns = ['converted_count', 'total_leads', 'conversion_rate', 'avg_score']
location_conv['conversion_rate'] = location_conv['conversion_rate'] * 100
location_conv = location_conv.sort_values('conversion_rate', ascending=False).head(10)

bars = ax2.barh(location_conv.index, location_conv['conversion_rate'], color='coral')
ax2.set_title('Top 10 Locations by Conversion Rate', fontsize=14, fontweight='bold')
ax2.set_xlabel('Conversion Rate (%)')
for i, (bar, value) in enumerate(zip(bars, location_conv['conversion_rate'])):
    ax2.text(value + 1, i, f'{value:.1f}%', va='center')

plt.tight_layout()
plt.savefig('../reports/industry_location_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved: reports/industry_location_analysis.png")

# Save industry analysis
industry_conv.to_csv('../reports/industry_analysis.csv')
print("Saved: reports/industry_analysis.csv")
```

```python
# Generate Executive Summary

print("\n" + "="*70)
print("Generating Executive Summary")
print("-"*70)

report_date = datetime.now().strftime('%Y-%m-%d')

exec_summary = {
    'Report Date': [report_date],
    'Total Leads': [len(df)],
    'High Priority Leads': [len(df[df['priority'] == 'HIGH'])],
    'Medium Priority Leads': [len(df[df['priority'] == 'MEDIUM'])],
    'Low Priority Leads': [len(df[df['priority'] == 'LOW'])],
    'Average Lead Score': [df['lead_score'].mean()],
    'Overall Conversion Rate (%)': [df['converted'].mean() * 100],
    'High Priority Conversion (%)': [df[df['priority'] == 'HIGH']['converted'].mean() * 100],
    'Total Converted Leads': [df['converted'].sum()],
    'Conversion Value (LKR)': [df['converted'].sum() * 50000],  # Assuming avg deal value
    'Top Industry': [df['industry'].mode()[0]],
    'Top Location': [df['location'].mode()[0]],
    'Avg Engagement Score': [df['engagement_score'].mean()]
}

exec_summary_df = pd.DataFrame(exec_summary)
exec_summary_df.to_csv(f'../reports/executive_summary_{report_date}.csv', index=False)

print(f"Saved: reports/executive_summary_{report_date}.csv")
print("\n Executive Summary:")
for key, value in exec_summary.items():
    if isinstance(value[0], float):
        print(f"   {key}: {value[0]:.2f}")
    else:
        print(f"   {key}: {value[0]}")
```

```python
# Generate Action Items Report

print("\n" + "="*70)
print("Generating Action Items Report")
print("-"*70)

# High priority leads requiring immediate action
high_priority = df[df['priority'] == 'HIGH'].nlargest(20, 'lead_score')

action_items = []
for idx, row in high_priority.iterrows():
    # Determine action based on lead characteristics
    if row['demo_requested'] == 'Yes':
        action = "Schedule demo ASAP - Lead requested demonstration"
        urgency = "URGENT"
    elif row['contact_level'] == 'C-Level':
        action = "Executive outreach - C-Level contact identified"
        urgency = "HIGH"
    elif row['engagement_score'] > 80:
        action = "Hot lead - High engagement, contact within 24h"
        urgency = "HIGH"
    else:
        action = "Standard follow-up within 24-48 hours"
        urgency = "MEDIUM"
    
    action_items.append({
        'Lead ID': row['lead_id'],
        'Company Size': row['company_size'],
        'Industry': row['industry'],
        'Location': row['location'],
        'Lead Score': row['lead_score'],
        'Priority': row['priority'],
        'Contact Level': row['contact_level'],
        'Urgency': urgency,
        'Recommended Action': action,
        'Assigned To': 'Senior Sales Rep' if urgency == 'URGENT' else 'Sales Team',
        'Pipeline Stage': row['pipeline_stage'],
        'Engagement Score': row['engagement_score']
    })

action_items_df = pd.DataFrame(action_items)
action_items_df.to_csv(f'../reports/action_items_{report_date}.csv', index=False)

print(f"Saved: reports/action_items_{report_date}.csv")
print(f"\n TOP 5 URGENT ACTIONS:")
urgent = action_items_df[action_items_df['Urgency'] == 'URGENT'].head(5)
for i, row in urgent.iterrows():
    print(f"\n   {i+1}. Lead #{row['Lead ID']} - {row['Industry']}")
    print(f"      Score: {row['Lead Score']:.1f}/100")
    print(f"      Action: {row['Recommended Action']}")
```

```python
# Generate Performance Metrics

print("\n" + "="*70)
print("Generating Performance Metrics")
print("-"*70)

metrics = {
    'Metric': [],
    'Value': [],
    'Target': [],
    'Status': []
}

metric_data = [
    ('High Priority Lead %', (len(df[df['priority'] == 'HIGH']) / len(df)) * 100, 20, '>='),
    ('Avg Lead Score', df['lead_score'].mean(), 50, '>='),
    ('Overall Conversion Rate %', df['converted'].mean() * 100, 5, '>='),
    ('High Priority Conversion %', df[df['priority'] == 'HIGH']['converted'].mean() * 100, 15, '>='),
    ('Avg Engagement Score', df['engagement_score'].mean(), 50, '>='),
    ('C-Level Contact %', (df['contact_level'].eq('C-Level').sum() / len(df)) * 100, 10, '>='),
    ('Demo Request Rate %', (df['demo_requested'].eq('Yes').sum() / len(df)) * 100, 25, '>=')
]

for metric_name, value, target, comparison in metric_data:
    if comparison == '>=':
        status = 'On Track' if value >= target else 'Below Target'
    else:
        status = 'On Track' if value <= target else 'Above Target'
    
    metrics['Metric'].append(metric_name)
    metrics['Value'].append(round(value, 2))
    metrics['Target'].append(target)
    metrics['Status'].append(status)

metrics_df = pd.DataFrame(metrics)
metrics_df.to_csv(f'../reports/performance_metrics_{report_date}.csv', index=False)

print(f"Saved: reports/performance_metrics_{report_date}.csv")
print("\n Performance Summary:")
print(metrics_df.to_string(index=False))
```

```python
# Generate Daily Summary Report

print("\n" + "="*70)
print("Generating Daily Summary Report")
print("-"*70)

daily_summary = {
    'Report Date': [report_date],
    'Total Leads': [len(df)],
    'High Priority': [len(df[df['priority'] == 'HIGH'])],
    'Medium Priority': [len(df[df['priority'] == 'MEDIUM'])],
    'Low Priority': [len(df[df['priority'] == 'LOW'])],
    'Converted': [df['converted'].sum()],
    'Conversion Rate': [f"{(df['converted'].sum() / len(df)) * 100:.2f}%"],
    'Avg Composite Score': [f"{df['lead_score'].mean():.2f}"],
    'Demos Requested': [len(df[df['demo_requested'] == 'Yes'])],
    'C-Level Contacts': [len(df[df['contact_level'] == 'C-Level'])]
}

daily_summary_df = pd.DataFrame(daily_summary)
daily_summary_df.to_csv(f'../reports/daily_summary_{report_date}.csv', index=False)

print(f"Saved: reports/daily_summary_{report_date}.csv")
```

```python
# Comprehensive Lead Report

print("\n" + "="*70)
print("Generating Comprehensive Lead Report")
print("-"*70)

comprehensive_report = df[['lead_id', 'company_size', 'industry', 'location',
                           'annual_revenue_lkr', 'engagement_score', 'contact_level',
                           'demo_requested', 'model_probability', 'lead_score',
                           'priority', 'pipeline_stage', 'converted']].copy()

# Add recommendations
def generate_recommendation(row):
    if row['priority'] == 'HIGH':
        if row['demo_requested'] == 'Yes':
            return 'URGENT: Follow up on demo request immediately'
        elif row['contact_level'] == 'C-Level':
            return 'HIGH VALUE: Schedule executive meeting'
        else:
            return 'PRIORITY: Initiate contact within 24 hours'
    elif row['priority'] == 'MEDIUM':
        return 'Follow up within 3-5 business days'
    else:
        return 'Nurture campaign - email sequence'

comprehensive_report['action_recommendation'] = comprehensive_report.apply(generate_recommendation, axis=1)

comprehensive_report.to_csv('../reports/comprehensive_lead_report.csv', index=False)
print("Saved: reports/comprehensive_lead_report.csv")
```

```python
# Key Insights

print("\n" + "="*70)
print("Key Insights & Recommendations")
print("="*70)

# Funnel insights
max_dropoff = funnel_df[funnel_df['drop_off_rate'] > 0]['drop_off_rate'].max()
if not pd.isna(max_dropoff):
    bottleneck_stage = funnel_df[funnel_df['drop_off_rate'] == max_dropoff]['stage'].values[0]
    print(f"\n  BIGGEST BOTTLENECK: {bottleneck_stage} stage ({max_dropoff:.1f}% drop-off)")
    print(f"    Recommendation: Focus resources on improving {bottleneck_stage} stage conversion")

best_conv_stage = funnel_df.loc[funnel_df['conversion_rate'].idxmax(), 'stage']
best_conv_rate = funnel_df['conversion_rate'].max()
print(f"\n BEST PERFORMING STAGE: {best_conv_stage} ({best_conv_rate:.1f}% conversion)")
print(f"    Recommendation: Analyze what's working in {best_conv_stage} stage")

# Priority insights
high_pct = (len(df[df['priority'] == 'HIGH']) / len(df)) * 100
print(f"\n HIGH PRIORITY LEADS: {high_pct:.1f}% of total")
if high_pct > 25:
    print(f"   Warning: High percentage of HIGH priority leads")
    print(f"    Recommendation: Review scoring thresholds")
else:
    print(f"  Good distribution - manageable for sales team")
    

# Conversion insights
overall_conv = df['converted'].mean() * 100
high_conv = df[df['priority'] == 'HIGH']['converted'].mean() * 100
print(f"\n CONVERSION RATES:")
print(f"   Overall: {overall_conv:.2f}%")
print(f"   HIGH Priority: {high_conv:.2f}%")
print(f"   Lift: {(high_conv / overall_conv - 1) * 100:.1f}%")
print(f"    Model successfully identifies high-conversion leads!")

print("\n" + "="*70)
print("ALL REPORTS GENERATED SUCCESSFULLY!")
print("="*70)

print(f"\n Generated Reports ({report_date}):")
print(f"    executive_summary_{report_date}.csv")
print(f"    action_items_{report_date}.csv")
print(f"    performance_metrics_{report_date}.csv")
print(f"    daily_summary_{report_date}.csv")
print(f"    comprehensive_lead_report.csv")
print(f"    funnel_metrics.csv")
print(f"    industry_analysis.csv")
print(f"    sales_funnel_visualization.png")
print(f"    industry_location_analysis.png")

print(f"\n Next Steps:")
print(f"   1. Review reports in reports/ folder")
print(f"   2. Launch dashboard: streamlit run dashboard.py")
print(f"   3. Share reports with sales team")
print(f"   4. Monitor high-priority leads daily")

print("\n" + "="*70)
print("FUNNEL TRACKING & REPORTING COMPLETE")
print("="*70)

```
